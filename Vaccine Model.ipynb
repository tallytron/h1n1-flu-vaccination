# Feature Engineering

```python
#converting age group into integer
vaccine.age_group[vaccine.age_group == '18 - 34 Years'] = 1
vaccine.age_group[vaccine.age_group == '35 - 44 Years'] = 2
vaccine.age_group[vaccine.age_group == '45 - 54 Years'] = 3
vaccine.age_group[vaccine.age_group == '55 - 64 Years'] = 4
vaccine.age_group[vaccine.age_group == '65+ Years'] = 5

vaccine.age_group = vaccine.age_group.astype(int)
```

```python
#converting education into integers as well
vaccine.education[vaccine.education == '< 12 Years'] = 1
vaccine.education[vaccine.education == '12 Years'] = 2
vaccine.education[vaccine.education == 'College Graduate'] = 3
vaccine.education[vaccine.education == 'Some College'] = 4

vaccine.education = vaccine.education.astype(int)
```

```python
#converting race into integers
vaccine.race[vaccine.race == 'Black'] = 1
vaccine.race[vaccine.race == 'Hispanic'] = 2
vaccine.race[vaccine.race == 'Other or Multiple'] = 3
vaccine.race[vaccine.race == 'White'] = 4

vaccine.race = vaccine.race.astype(int)
```

```python
#sex as integers
vaccine.sex[vaccine.sex == 'Female'] = 0
vaccine.sex[vaccine.sex == 'Male'] = 1

vaccine.sex = vaccine.sex.astype(int)
```

```python
#converting marital status into numerical and integers
vaccine.marital_status[vaccine.marital_status == 'Married'] = 0
vaccine.marital_status[vaccine.marital_status == 'Not Married'] = 1

vaccine.marital_status = vaccine.marital_status.astype(int)
```

```python
#converting rent or own into numerical and integers
vaccine.rent_or_own[vaccine.rent_or_own == 'Rent'] = 0
vaccine.rent_or_own[vaccine.rent_or_own == 'Own'] = 1

vaccine.rent_or_own = vaccine.rent_or_own.astype(int)
```

```python
#converting employment status into numerical and integers
vaccine.employment_status[vaccine.employment_status == 'Employed'] = 1
vaccine.employment_status[vaccine.employment_status == 'Not in Labor Force'] = 2
vaccine.employment_status[vaccine.employment_status == 'Unemployed'] = 3

vaccine.employment_status = vaccine.employment_status.astype(int)
```

```python
#converting census msa into numerical and integers
vaccine.census_msa[vaccine.census_msa == 'MSA, Not Principle  City'] = 1
vaccine.census_msa[vaccine.census_msa == 'MSA, Principle City'] = 2
vaccine.census_msa[vaccine.census_msa == 'Non-MSA'] = 3

vaccine.census_msa = vaccine.census_msa.astype(int)
```

```python
def cv_results_to_df(cv_results):
    results = pd.DataFrame(list(cv_results['params']))
    results['mean_fit_time'] = cv_results['mean_fit_time']
    results['mean_score_time'] = cv_results['mean_score_time']
    results['mean_train_score'] = cv_results['mean_train_score']
    results['std_train_score'] = cv_results['std_train_score']
    results['mean_test_score'] = cv_results['mean_test_score']
    results['std_test_score'] = cv_results['std_test_score']
    results['rank_test_score'] = cv_results['rank_test_score']

    results = results.sort_values(['mean_test_score'], ascending=False)
    return results
```
